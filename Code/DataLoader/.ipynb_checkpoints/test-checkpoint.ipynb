```python
from transformers import BertTokenizer
```

```python
tokenizer_dir = "bert-SNP"
tokenizer.save(tokenizer_dir)

max_len=4096

PAD = "$"
MASK = "."
UNK = "?"
SEP = "|"
CLS = "*"

bert_tokenizer = BertTokenizer.from_pretrained(tokenizer_dir,
                                                    do_lower_case=True,
                                                    do_basic_tokenize=True,
                                                    tokenize_chinese_chars=True,
                                                    pad_token=PAD,
                                                    mask_token=MASK,
                                                    unk_token=UNK,
                                                    sep_token=SEP,
                                                    cls_token=CLS,
                                                    padding_side="right"
                                                    )
```

```python
import numpy as np
```

```python
seq_dir = "sample_30000.txt"
with open(seq_dir, 'r',encoding='utf-8') as file:
    seq_list = []
    for line in file:
        line1=line.strip()
        if line1!='':
            seq_list.append(line1)
```

```python
seq0=seq_list[5]
retval = bert_tokenizer.encode(seq0,truncation=True, max_length=max_len)
```
